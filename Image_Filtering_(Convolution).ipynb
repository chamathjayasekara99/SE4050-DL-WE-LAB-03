```python
import cv2
from matplotlib import pyplot as plt
import numpy as np
```

### Load the image using OpenCV

```python
im = cv2.imread('/content/sample_data/lenna.png') #change the path as needed
```

```python
im.shape
```

### Convert the RGB image into a grayscale image

```python
im = cv2.cvtColor(im, cv2.COLOR_BGR2GRAY)
im.shape
```

### Define vertical and horizontal filter kernels

```python
v_kernel = np.array([
    [-1, 0, 1],
    [-1, 0, 1],
    [-1, 0, 1]
])
```

```python
h_kernel = np.array([
    [-2, -2, -2],
    [0, 0, 0],
    [1, 1, 1]
])
```

```python
main_diagonal_kernel = np.array([
    [1,0,-1],
    [0,0,0],
    [2,0,2]
])
```

```python
secondary_diagonal_kernel = np.array([
    [-2,0,2],
    [0,0,0],
    [-1,0,1]
])
```

### Do the filtering

```python
v_image = cv2.filter2D(im, -1, v_kernel)
h_image = cv2.filter2D(im, -1, h_kernel)
m_d_image = cv2.filter2D(im,-1,main_diagonal_kernel)
s_d_image =cv2.filter2D(im,-1,secondary_diagonal_kernel)
```

### Display results

```python
plt.imshow(im, cmap='gray')
plt.title('Original Image')

plt.figure()
plt.imshow(v_image, cmap='gray')
plt.title('Vertical Image')

plt.figure()
plt.imshow(h_image, cmap='gray')
plt.title('Horizontal Image')

plt.figure()
plt.imshow(m_d_image, cmap='gray')
plt.title('Main Diagonal Image')

plt.figure()
plt.imshow(s_d_image, cmap='gray')
plt.title('Secondary Diagonal Image')
```
